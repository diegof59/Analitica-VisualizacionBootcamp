```python
columns_of_interest = [
    '2022 Population', '2020 Population', '2015 Population', '2010 Population',
    '2000 Population', '1990 Population', '1980 Population', '1970 Population',
    'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage'
]

df_filtered = df[columns_of_interest]

corr_matrix = df_filtered.corr()
```

```python
# Filas que contienen valores no numéricos en '2022 Population'
non_numeric_rows = df[pd.to_numeric(df['2022 Population'], errors='coerce').isna()]

# Mostrar las filas con valores no numéricos
print(non_numeric_rows)

# Crear una máscara que identifique las filas donde '2022 Population' no es numérico
mask_non_numeric = df['2022 Population'].apply(lambda x: not str(x).replace('.', '', 1).isdigit())

# Mostrar las filas donde el valor de '2022 Population' no es numérico
non_numeric_values = df[mask_non_numeric]

# Ver las filas con valores no numéricos en '2022 Population'
print(non_numeric_values[['Country', '2022 Population']])
```
